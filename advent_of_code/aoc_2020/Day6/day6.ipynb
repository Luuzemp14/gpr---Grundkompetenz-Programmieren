```python
# Part 1

def handle_1(lst):
    print(lst)
def handle_file(file_name, handle):
    with open(file_name) as f:
        lst = []
        for line in f:
            if line != "\n":
                lst.append(line.strip())
            else:
                handle(lst)
                lst = []
        if len(lst) != 0:
            handle(lst)
        
handle_file("input_day6.txt", handle_1)
```
